Here take 0,1 values for showing male and female and then remove any one beacause we can identify it easily like same goes for embarked and pclass

```python
pd.get_dummies(df2['Sex']).head()
```

```python
Sex = pd.get_dummies(df2['Sex'],drop_first = True)
Sex.head()
```

```python
E = pd.get_dummies(df2['Embarked'],drop_first = True)
E.head()
```

```python
P = pd.get_dummies(df2['Pclass'],drop_first = True)
P.head()
```

```python
df2 = pd.concat([df2,Sex,E,P],axis=1)
df2.head()
```

```python
df2.drop(['Name','PassengerId','Sex','Pclass','Ticket','Embarked'],axis = 1,inplace = True)
```

```python
df2.head()
```

TRAIN DATA

```python
x = df2.drop('Survived',axis = 1)
y = df2['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=1)
```

LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
pk = LogisticRegression()
```

```python
pk.fit(x_train,y_train)
```

```python
prediction = pk.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,prediction)
```

ACCURACY

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prediction)
```
